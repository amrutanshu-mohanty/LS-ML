### Cartoon
For the bart image, you need to mask the dress. Write all your code in the cell below and the final image should be stored in ```cartoon_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)
green = [0, 255, 50] # green in bgr
green = np.uint8([[green]]) # since it usually converts images, cv2.cvtColor needs 3D array
hsv_green = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
hsv_image = cv2.cvtColor(cartoon, cv2.COLOR_BGR2HSV)
lowerLimit = hsv_green[0][0][0] - 10, 100, 100
upperLimit = hsv_green[0][0][0] + 10, 255, 255
"""
Referring to Copilot to understand the limits:
Hue (H): This represents the color itself. For green, the typical range is around 60 to 180 (out of 0-360). So, hsv_green[0][0][0] - 10 would be the lower limit for green hue, and hsv_green[0][0][0] + 10 would be the upper limit. The subtraction and addition of 10 allow for a slight variation in the green hue.
Saturation (S): This measures the intensity or purity of the color. A higher saturation value means a more vivid color. For green, a saturation range of 100 to 255 is often used. So, your lower limit is 100, and the upper limit is 255.
Value (V): This represents the brightness. For green, you can set a wide range, such as 100 to 255. This allows you to detect green objects regardless of their brightness.
"""
lowerLimit = np.array(lowerLimit, dtype=np.uint8)
upperLimit = np.array(upperLimit, dtype=np.uint8)

mask = cv2.inRange(hsv_image, lowerLimit, upperLimit)   # simply takes other values to 0
"""
cv2.imshow('mask', mask)
cv2.waitKey(0)
"""
# bgr_image = cv2.cvtColor(mask, cv2.COLOR_HSV2BGR)

trial = hsv_image.copy()
for x in trial:
    for y in x:
        for i in range(3):
            if y[i] < lowerLimit[i] or y[i] > upperLimit[i]:
                for j in range(3):
                    y[j] = 0

"""
cv2.imshow('trial', trial)
cv2.waitKey(0)
"""

cartoon_y_mentee = cv2.cvtColor(trial, cv2.COLOR_HSV2BGR)   # (in BGR format only)
cv2.imshow('final', cartoon_y_mentee)
cv2.waitKey(0)
# END CODE HERE (the final processed image should be cartoon_y_mentee)
````

### Girl
For the girl image, you need to detect the edges. Write all your code in the cell below and the final image should be stored in ```girl_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)

girl_y_mentee = cv2.medianBlur(girl, 5)
girl_y_mentee = cv2.Canny(girl_y_mentee, 80, 175)

cv2.imshow('girl_y_mentee', girl_y_mentee)
cv2.imshow('girl_y', girl_y)
cv2.waitKey(0)

#END CODE HERE (the final processed image should be girl_y_mentee)
````

### Fingerprint
For the fingerprint image, you need to convert it to binary image. Binary images have 2 intensities only (0 or 255). Write all your code in the cell below and the final image should be stored in ```cartoon_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)

fingerprint_y_mentee = None #(in Binary format only i.e. pixel intensity = 0 or 255)
fingerprint_y_mentee = cv2.cvtColor(fingerprint, cv2.COLOR_BGR2GRAY)
# now we can see the color is actually inverted, so we subtract from 255
fingerprint_y_mentee = 255 - fingerprint_y_mentee
ret, fingerprint_y_mentee = cv2.threshold(fingerprint_y_mentee, 150, 255, cv2.THRESH_BINARY)
cv2.imshow('girl_y_mentee', fingerprint_y_mentee)
cv2.imshow('girl_y', fingerprint_y)
cv2.waitKey(0)
#END CODE HERE (the final processed image should be fingerprint_y_mentee)
````
